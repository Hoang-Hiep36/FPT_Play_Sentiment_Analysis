```python
import pymongo
from pymongo import MongoClient
```

```python
CONNECTION_STRING = "mongodb+srv://..."
```

```python
client = MongoClient(CONNECTION_STRING)
```

```python
db = client['...']
```

```python
collection = db["Comments"]
```

```python
cursor = collection.find().limit(50000)
```

```python
list_cur = list(cursor)
```

```python
import pandas as pd
df = pd.DataFrame(list_cur)
```

```python
df
```

```python
df['content'].str.split(' ')
```

```python
comments_lite = df['content']
```

```python
word_list = [word for phrase in comments_lite.str.cat(sep=' ').split(' ') for word in phrase.split()]
```

```python
word_list
```

```python
from collections import Counter
```

```python
pd.Series(word_list).value_counts()[:1000]
```

```python
# dem so lan xuat hien cua moi tu
ketqua = Counter(word_list)
```

```python
df_ketqua = pd.DataFrame([ketqua])
```

```python
df_ketqua
```

```python
df_ketqua.transpose().to_excel('top_key_words.xlsx')
```
